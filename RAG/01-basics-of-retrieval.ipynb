## Preprocessing

```python
def preprocess_text(text, include_stopwords=False, include_boolean_stopwords=False):
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text.lower())
    # Remove stop words and non-alphanumeric characters
    if include_stopwords:
        filtered_tokens = [word for word in tokens if word.isalnum()]
    else:
        stop_words = set(nltk.corpus.stopwords.words('english'))
        if include_boolean_stopwords:
            stop_words = stop_words - {'and', 'not', 'or'}
        filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return ' '.join(filtered_tokens)
```

```python
preprocessed_documents = [preprocess_text(doc) for doc in documents]
preprocessed_documents[:3]
```

## Retrieve with Vector Space Model

### TF-IDF
TF(Term Frequency) = times term T appears in  document D / total number of terms in document D

IDF(Inverse Document Frequence) = log(number of documents / number of documents containing term T)

TF-IDF = TF x IDF

If some term appears in every document, then TF-IDF is 0. (This kind of words are treated as stopwords, and often removed in preprocessing.)

### Pros-Cons
Pros: Ranking, Partial matches
Cons: Intensive for large datasets

```python
# TF-IDF Vectorizor
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(preprocessed_documents)
# 100 is the number of documents, 325 is the number of dimensions
tfidf_matrix.shape
```

```python
# Query
def search_tf_idf(query):
    preprocessed_query = preprocess_text(query)
    query_vector = vectorizer.transform([preprocessed_query])
    # Calculate cosine similarity
    cosine_similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()
    # Get the top 5 most similar documents
    top_indices = cosine_similarities.argsort()[-5:][::-1]

    # Print the top 5 most similar documents
    for i in top_indices:
        print(f"Document: {documents[i]}")
        print(f"Cosine Similarity: {cosine_similarities[i]}")
        print()

query = "Sailing in Croatia"
search_tf_idf(query)
```

## Retrieve with Boolean Conditions
Pros: Simple, Precision

Cons: No Ranking, Binary result, Not fit for complex queries

```python
import os
import shutil
from whoosh.index import create_in, open_dir
from whoosh.fields import Schema, TEXT, ID
from whoosh.qparser import QueryParser
```

```python
# Preprocessing
# preprocessed_documents = [preprocess_text(doc, include_boolean_stopwords=True) for doc in documents]
# Create a schema for the index
if os.path.exists("01_index"):
    shutil.rmtree("01_index")
os.mkdir("01_index")
schema = Schema(title = ID(stored=True,unique=True),content=TEXT(stored=True))
# Create an index
ix = create_in("01_index", schema)
# Add documents to the index
writer = ix.writer()
for i, doc in enumerate(documents):
    writer.add_document(title=str(i),content=doc)
writer.commit()
```

```python
# Querying
def search_boolean(query):
    ix = open_dir("01_index")
    with ix.searcher() as searcher:
        query_parser = QueryParser("content", ix.schema)
        parsed_query = query_parser.parse(query)
        results = searcher.search(parsed_query, limit=5)
        print(f"query  = {query}")
        for result in results:
            print(f"Document: {result['content']}")
query = "sailing NOT croatia"
search_boolean(query)
query = "sailing AND Croatia"
search_boolean(query)
query = "unesco OR Croatia"
search_boolean(query)
```

## Retrieve with Probabilistic Model
Pros: Ranking, Probabilistic Reasoning, Adaptability
Cons: Complex, Need initial Assumptions, Dependent to data

```python
from rank_bm25 import BM25Okapi

def preprocess_text_and_get_tokens(text, include_stopwords=False, include_boolean_stopwords=False):
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text.lower())
    # Remove stop words and non-alphanumeric characters
    if include_stopwords:
        filtered_tokens = [word for word in tokens if word.isalnum()]
    else:
        stop_words = set(nltk.corpus.stopwords.words('english'))
        if include_boolean_stopwords:
            stop_words = stop_words - {'and', 'not', 'or'}
        filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return filtered_tokens

# Preprocess the documents for BM25
preprocessed_tokens = [preprocess_text_and_get_tokens(doc, include_stopwords=True) for doc in documents]

# Initialize BM25
bm25 = BM25Okapi(preprocessed_tokens)

# Query
def search_bm25(query):
    query_tokens = preprocess_text_and_get_tokens(query, include_stopwords=True)
    scores = bm25.get_scores(query_tokens)
    top_indices = np.argsort(scores)[-5:][::-1]
    
    print(f"query  = {query}")
    for i in top_indices:
        print(f"Document: {documents[i]}")
        print(f"BM25 Score: {scores[i]}")
        print()
query = "Sailing in UNESCO"
search_bm25(query)
```
